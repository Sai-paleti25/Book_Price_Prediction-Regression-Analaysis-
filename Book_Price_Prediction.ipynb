#  Light GBM Regressor

#### Light GBM use histogram based algorithm i.e it buckets continuous feature values into discrete bins which fasten the training procedure.
#### It produces much more complex trees by following leaf wise split approach rather than a level-wise approach which is the main factor in achieving higher accuracy. However, it can sometimes lead to overfitting which can be avoided by setting the max_depth parameter.

```python
import lightgbm as lgb
```

```python
#convert the dataset to lgbDataset
df_train = lgb.Dataset(X_train,y_train)
```

```python
hyper_params = {
    'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': ['l1', 'auc'],
    'learning_rate': 0.05,
    'num_iterations':1000,
    'num_leaves':150,
    'max_depth':7,
    'max_bin':200
}
```

```python
rg = lgb.LGBMRegressor(**hyper_params)
```

```python
rg.fit(X_train,y_train)
```

```python
y_pred_train = rg.predict(X_train)
```

```python
print("Train MSE:",np.sqrt(mean_squared_error(y_train,y_pred_train)))
```

```python
print("Train r2_score:",r2_score(y_train, y_pred_train))
```

```python
# Predicting the Test set results
y_pred = rg.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
r2_score(y_test, y_pred)
```

```python
print("Test MSE:",np.sqrt(mean_squared_error(y_test,y_pred)))
```
